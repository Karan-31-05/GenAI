```python
# STEP 1: Install dependencies
!pip install langchain google-generativeai langchain-google-genai youtube-transcript-api sentence-transformers faiss-cpu langchain-community

# STEP 2: Set your Gemini API Key
import os
os.environ["GOOGLE_API_KEY"] = "AIzaSyAh5KFurJrIwvNE6kdqGkQp8XCL_E7XFwk"  # 🔑 replace this with your actual key

# STEP 3: Import modules
from youtube_transcript_api import YouTubeTranscriptApi
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import FAISS
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.chains import RetrievalQA

# STEP 4: Function to get transcript
def get_youtube_transcript(video_url):
    if "youtu.be/" in video_url:
        video_id = video_url.split("youtu.be/")[-1].split("?")[0]
    elif "watch?v=" in video_url:
        video_id = video_url.split("watch?v=")[-1].split("&")[0]
    else:
        raise ValueError("Invalid YouTube URL format.")

    transcript = YouTubeTranscriptApi.get_transcript(video_id)
    full_text = " ".join([t['text'] for t in transcript])
    return full_text

# STEP 5: Create retriever from transcript
def create_retriever_from_text(text):
    text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=100)
    docs = text_splitter.create_documents([text])
    embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
    vectorstore = FAISS.from_documents(docs, embeddings)
    retriever = vectorstore.as_retriever()
    return retriever

# STEP 6: Create RAG chain with Gemini
def get_summary_rag_chain(retriever):
    llm = ChatGoogleGenerativeAI(model="models/gemini-1.5-pro", temperature=0.5)  # ✅ FIXED model name
    rag_chain = RetrievalQA.from_chain_type(
        llm=llm,
        retriever=retriever,
        return_source_documents=False
    )
    return rag_chain

# STEP 7: Main function
def summarize_youtube_video(url):
    transcript = get_youtube_transcript(url)
    retriever = create_retriever_from_text(transcript)
    rag_chain = get_summary_rag_chain(retriever)

    question = "Give me a summary of the video in simple language."
    summary = rag_chain.run(question)
    return summary

# STEP 8: Input + Run
video_url = input("Paste YouTube video URL: ")
print("\n⏳ Summarizing... please wait...\n")
summary = summarize_youtube_video(video_url)
print("✅ Summary:\n")
print(summary)
```
